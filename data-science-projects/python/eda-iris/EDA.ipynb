# Iris EDA

This notebook performs a quick EDA on `iris.csv` using **pandas** and **matplotlib**.

Steps:
1. Load data
2. Inspect structure and summary stats
3. Plot histograms and pairwise scatter


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('iris.csv')
df.head()
```

```python
# Basic structure and summary
display(df.info())
df.describe(include='all')
```

```python
# Null checks
df.isna().sum()
```

```python
# Histograms (one chart per feature)
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    plt.figure()
    df[col].hist()
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Pairwise scatter (simple)
cols = list(df.select_dtypes(include='number').columns)
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        plt.figure()
        plt.scatter(df[cols[i]], df[cols[j]])
        plt.xlabel(cols[i])
        plt.ylabel(cols[j])
        plt.title(f'{cols[i]} vs {cols[j]}')
        plt.show()
```
